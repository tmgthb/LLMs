```python
!pip install --upgrade openai
import os
from pathlib import Path
from IPython.display import Audio
from openai import OpenAI 
client = OpenAI(
    api_key=os.getenv("openaikey"))
```

```python
response = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What you think the person in the image is doing?"},
                {
                    "type": "image_url",
                    "image_url": "https://img-s-msn-com.akamaized.net/tenant/amp/entityid/AA18Lnc8.img?w=1920&h=1080&q=60&m=2&f=jpg",
                },
            ],
        }
    ],
    max_tokens=150,
)
text_to_speech = response.choices[0].message.content
```

```python
speech_file_path =  "C:/Users/Teemu Määttä/Desktop/Python/speech.mp3"
response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input=text_to_speech
)
```

```python
response.stream_to_file(speech_file_path)
Audio(speech_file_path)
```
